```python
import pandas as pd
```

```python
df = pd.read_csv('/content/links.csv')
df = df[['Source', 'Target']]
df
```

```python
from dataclasses import dataclass, field

@dataclass
class Site:
  name: str
  hosts: set = field(default_factory=set)
```

```python
sites_list = []
site_counter = 1

for row in df.itertuples():
  print(row.Source, row.Target)
  site_exists = False
  if len(sites_list) > 0:
    for site in sites_list:
      if row.Source in site.hosts or row.Target in site.hosts:
        print(f'{row.Source} is in {site.hosts}')
        site.hosts.add(row.Source)
        site.hosts.add(row.Target)
        site_exists = True

  if not site_exists:
    print(f'Creating site {site_counter}')
    site = Site(site_counter)
    site.hosts.add(row.Source)
    site.hosts.add(row.Target)
    print(site.name, site.hosts)
    sites_list.append(site)
    site_counter += 1



```

```python
sites_list = []
site_counter = 1

for row in df.itertuples():
  site_exists = False

  if len(sites_list) > 0:
    for site in sites_list:
      if row.Source in site.hosts or row.Target in site.hosts:
        site.hosts.add(row.Source)
        site.hosts.add(row.Target)
        site_exists = True

  if not site_exists:
    print(f'Creating site {site_counter}')
    new_site = Site(site_counter)
    source = row.Source

    res = df.query("Source == @source or Target == @source")
    hosts = res['Source'].to_list() + res['Target'].to_list()
    for host in hosts:
      new_site.hosts.add(host)

    target = row.Target
    res = df.query("Source == @target or Target == @target")
    hosts = res['Source'].to_list() + res['Target'].to_list()
    for host in hosts:
      new_site.hosts.add(host)

    print(f"site {new_site.name} has hosts: {new_site.hosts}")
    sites_list.append(new_site)
    site_counter += 1
```
